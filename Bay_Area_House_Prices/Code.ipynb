```python
y=df_["lastsoldprice"]
df_.drop("lastsoldprice",inplace=True,axis=1)
df_.drop("address",inplace=True,axis=1)


```

```python
x=df_[["bathrooms","bedrooms","finishedsqft","lastsolddate","neighborhood","totalrooms","latitude","longitude","usecode","yearbuilt","zindexvalue"]]

x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.32, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

clf = RandomForestRegressor(random_state=42,n_estimators=30,criterion="mae",min_samples_split=2,n_jobs=-1)
clf.fit(X_train, y_train)
clf.score(X_test,y_test)
```
